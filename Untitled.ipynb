```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("lol_team_clean.csv")

# 只保留需要的列 & 丢掉缺失
cols = ["win", "golddiffat10", "kills_diff_10", "firstdragon"]
df = df[cols].dropna()

# 构造特征矩阵 X，注意把 gold 缩放一下（方便系数）
X = pd.DataFrame({
    "gold10k": df["golddiffat10"] / 1000.0,   # 以1k gold为单位
    "killsDiff10": df["kills_diff_10"],
    "firstDragon": df["firstdragon"]          # 假设 1=我方拿到，0=没拿/对面拿
})

y = df["win"]  # 0/1

model = LogisticRegression()
model.fit(X, y)

print("intercept_", model.intercept_)   # β0
print("coef_", model.coef_)            # [β1, β2, β3]
```
